```python
import pandas as pd
```

```python
df_contractors = pd.read_excel('./Contractor Onboarding Data_01.01.16 - 09.04.19.xlsx',header=14,na_values='-')
```

```python
df_contractors['year'] = pd.to_datetime(df_contractors['Account Creation Date']).dt.year
```

```python
import random
import numpy as np

def my_train_test_split(df,train_size=0.8):
    index_list = list(df.index)
    random.shuffle(index_list)
    train_index = index_list[0:int(len(index_list)*train_size)]
    df['train_test'] = np.array([0]*len(index_list))
    df.loc[train_index,'train_test'] = 1
    return df
```

```python
df_contain = pd.DataFrame()
for i in [2016,2017,2018,2019]:
    df = df_contractors.loc[(df_contractors['year'] == i),['Dealer ID','year']]
    df = my_train_test_split(df)
    df_contain = pd.concat([df_contain,df],axis=0)
```
